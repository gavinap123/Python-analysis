```python
import seaborn as sns
```

```python
mpg = sns.load_dataset('mpg')
```

```python
mpg.head()
```

```python
#Create a visual showing the relationship between horsepower and fuel economy.
sns.relplot(data=mpg, y='mpg', x='horsepower')
```

```python
#Create a table summarizing fuel economy by country of origin
mpg_by_origin = mpg.groupby('origin')['mpg'].mean().reset_index()
print(mpg_by_origin)
```

```python
#Create a visual showing how fuel economy has changed over time
mean_mpg_by_year = mpg.groupby('model_year')['mpg'].mean().reset_index()
sns.lineplot(x='model_year', y='mpg', data=mean_mpg_by_year, marker='o')
```

```python
#Create two other visuals or tables related to fuel economy.
```

```python
# Create a visual showing the acceleration to cylinders relationship by orgin.
mean_acc_cyl_origin = mpg.groupby(['origin', 'cylinders'])['acceleration'].mean().reset_index()
print(mean_acc_cyl_origin)
```

```python
sns.lineplot(x='cylinders', y='acceleration', hue='origin', data=mean_acc_cyl_origin)
```

```python
#Create a visual showing the relationship between weight and fuel economy.
sns.relplot(data=mpg, y='mpg', x='weight')
```
